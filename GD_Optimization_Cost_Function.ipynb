```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)

m = 100
w = 4

noise = np.random.randn(m) * 1.5

x = np.linspace(0, 10, m)

y = w * x + noise
```

```python
plt.scatter(x, y, label='Data', color='blue', marker='o')


plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Trend')
plt.legend()

plt.show()
```

```python
learning_rate = 0.0001
numIter = 1000
initialVal = 4.002
```

```python
def compute_cost(w, x, y):
    prediction = w * x
    m = len(y) #number of training samples
    cost = (1/(2*m))*np.sum(prediction - y) **2
    return cost
```

```python
def gradient_descent(x, y, learning_rate, num_iterations):
    w = initialVal
    m = len(x)
    
    for i in range(num_iterations):
        predictions = w * x
        gradient = (1 / m) * np.dot(predictions - y, x)
        w -= learning_rate * gradient
        cost = compute_cost(w, x, y)
        
        if i % 100 == 0:
            print(f"Iteration {i}: Cost = {cost}, w = {w}")
    
    return w
```

```python
optimal_w = gradient_descent(x, y, learning_rate, numIter)
print(f"Optimal w after gradient descent: {optimal_w}")
```

```python
plt.scatter(x, y, label='Data', color='blue', marker='o')

plt.plot(x, optimal_w * x, color='red', label='Linear Regression Line')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```
